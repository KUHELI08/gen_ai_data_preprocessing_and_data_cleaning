```python
from langchain import HuggingFacePipeline
from transformers import AutoTokenizer, pipeline
import transformers
import torch

```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
model="daryl149/llama-2-7b-chat-hf"
```

```python
tokenizer=AutoTokenizer.from_pretrained(model)
```

```python
pipeline=transformers.pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    torch_dtype=torch.bfloat16,
    device_map="auto",
    max_length=1000,
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id
)
```

```python
llm=HuggingFacePipeline(pipeline=pipeline, model_kwargs={'temperature':0})
```

```python
prompt="What would be a good name for a company that makes colorful socks"
```

```python
print(llm(prompt))
```

```python
from langchain import PromptTemplate
from langchain.chains import LLMChain
```

```python
prompt_template1=PromptTemplate(input_variables=["cusine"],template="I want to open a restaurant for {cusine} food. Suggest a fency name for this")
```

```python
input_prompt=prompt_template1.format(cusine="Italian")
```

```python
print(input_prompt)
```

```python
chain = LLMChain(llm=llm, prompt=prompt_template1, verbose=True)
response= chain.run("Italian")
print(response)
```
